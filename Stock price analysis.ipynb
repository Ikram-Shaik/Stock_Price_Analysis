```python
# Importing Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing all the required datasets

```python
BAC=pd.read_csv("BAC.csv")
```

```python
BAC["Stock"]="Bank of America Corporation"
```

```python
C=pd.read_csv("C.csv")
```

```python
C["Stock"]="Citigroup"
```

```python
GS=pd.read_csv("GS.csv")
```

```python
GS["Stock"]="Goldman Sachs"
```

```python
JPM=pd.read_csv("JPM.csv")
```

```python
JPM["Stock"]="JPMorgan"
```

```python
MS=pd.read_csv("MS.csv")
```

```python
MS["Stock"]="Morgan Stanley"
```

```python
WFC=pd.read_csv("WFC.csv")
```

```python
WFC["Stock"]="Wells Fargo & Company"
```

## Convering all the datasets into a single DataFrame

```python
df=pd.concat([BAC,C,GS,JPM,MS,WFC])
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.Date.dtype
```

```python
# Converting Date column datatype to datetime
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.head()
```

```python
df.Date.dtype
```

```python
df.dtypes
```

```python
# Adding a new column name Year
df["Year"]=df["Date"].dt.year
```

```python
df.head()
```

## 1. Scatter plot

```python
sns.scatterplot(x="Close",y="Volume",data=df,marker="x",hue="Stock",style="Stock")
plt.legend(bbox_to_anchor=(1,1),loc="upper left")
plt.title("Closing Price vs Volume")
plt.xlabel("Closing Price")
plt.ylabel("Volume")
plt.show()
```

### Insights Found:
Stock with low Prices have higher trading volume  --- Bank of America Corporation has high trading volume and stock values are less
<br>
Stock with high Prices have low trading volume  --- Citigroup has high trading volume and stock values are less

## 2. Barplot

```python
df.head(1)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="Year",y="Volume",data=df,palette="rainbow_r",hue="Year",legend=False)
plt.title("Volume of Stocks vs Year")
plt.show()
```

The volume of stocks traded in 2009 is very high compared to other years

```python
plt.figure(figsize=(12,5))
sns.barplot(x="Year",y="Volume",data=df,hue="Stock")
plt.title("Volume of various stocks vs Year")
plt.show()
```

Bank of America Corporation has the the volume of stocks every Year

## 3. Histogram

```python
sns.histplot(x="Open",data=df,bins=30,kde=True)
plt.title("Distribution of Opening Prices of Stock")
plt.xlabel("Opening Price")
plt.show()
```

### Insights:
The price of most of the opening stock range between 20-40

## 4. Displot

```python
sns.displot(x="Close",kind="kde",data=df)
plt.title("Distribution of Closing Prices of Stocks")
plt.xlabel("Closing Prices")
plt.show()
```

### Insights:
The closing price of the stock mostly ranges between 0-100

## 5. Pair plot

```python
sns.pairplot(data=df[["Open","High","Low","Close"]])
plt.show()
```

## 6. Box Plot

```python
sns.catplot(x="Year",y="Close",data=df,kind="box",height=6,aspect=2,palette="rainbow_r",hue="Year",legend=False)
plt.title("Box Plot For Year vs Closing Price Of Stocks")
plt.xlabel("Closing Price")
plt.show()
```

## 7. Violin Plot

```python
sns.catplot(x="Year",y="High",data=df,kind="violin",height=6,aspect=2,palette="rainbow_r",hue="Year",legend=False)
plt.show()
```
